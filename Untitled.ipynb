```python
import requests
import zipfile
import io
import os

repo_url = "https://github.com/suhasdevmane/talk2abacws-survey"
zip_url = repo_url.rstrip('/') + "/archive/refs/heads/main.zip"  # Most repos use 'main' branch

response = requests.get(zip_url)
response.raise_for_status()

with zipfile.ZipFile(io.BytesIO(response.content)) as z:
    # Extract all to the current directory, remove the top-level folder in the zip
    for member in z.namelist():
        # Remove the first folder ("talk2abacws-survey-main/")
        path_parts = member.split('/', 1)
        if len(path_parts) < 2 or not path_parts[1]:
            continue  # skip top-level folder entries
        target_path = path_parts[1]
        # Extract file or directory
        if not os.path.exists(os.path.dirname(target_path)) and os.path.dirname(target_path):
            os.makedirs(os.path.dirname(target_path), exist_ok=True)
        if member[-1] != '/':  # skip directories, handle files only
            with z.open(member) as source_file, open(target_path, "wb") as target_file:
                target_file.write(source_file.read())
            # Make all files readable and writable for everyone (no sudo needed)
            os.chmod(target_path, 0o666)
print("Download and extraction complete.")
```
